# Step 1: Find and download a dataset

Data from https://www.kaggle.com/datasets/mathchi/diabetes-data-set

This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases.

Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

- **Pregnancies**: Number of times pregnant
- **Glucose**: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
- **BloodPressure**: Diastolic blood pressure (mm Hg)
- **SkinThickness**: Triceps skin fold thickness (mm)
- **Insulin**: 2-Hour serum insulin (mu U/ml)
- **BMI**: Body mass index (weight in kg/(height in m)^2)
- **DiabetesPedigreeFunction**: Diabetes pedigree function
- **Age**: Age (years)
- **Outcome**: Class variable (0 or 1)

### Import Packages & Data

```python
import numpy as np
import pandas as pd 

original_data = pd.read_csv("data/diabetes.csv")
```

# Step 2: Divide into a training set and a testing set

### Stratified Shuffle Splitting

```python
# import package
from sklearn.model_selection import StratifiedShuffleSplit


# declare our X inputs and y outcomes
X = original_data.drop("Outcome", axis=1)
y = original_data["Outcome"]


# create instance of splitter object
splitter = StratifiedShuffleSplit(n_splits=5,
                               test_size=0.2, 
                               random_state=42)

# use the splitter to split the X and y sets
for train_index, test_index in splitter.split(X, y):
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]

    
    
# Prints ======================

print("X_train.shape = ", X_train.shape)
print("X_test.shape = ", X_test.shape)

print("y_train.shape = ", y_train.shape)
print("y_test.shape = ", y_test.shape)

print("\n")
print("y_train class proportions: \n", y_train.value_counts(normalize=True))

print("\n")
print("y_test class proportions: \n", y_test.value_counts(normalize=True))
```

### Clean Missing Data

**In the visualization section, I discovered there are some cases where Glucose, BMI, Blood Pressure are 0, which is not possible (assuming all sampled individuals were alive)! They need to be cleaned.**

```python
# See how many cases have 0 Glucose, BMI, or Blood Pressure

zero_g = training_set["Glucose"] == 0
zero_bmi = training_set["BMI"] == 0
zero_bp = training_set["BloodPressure"] == 0

all_zero_cases = training_set[zero_g | zero_bmi | zero_bp]

print(len(all_zero_cases), "cases with 0 Glucose, BMI, or Blood Pressure")
```

**There are also cases with skin thickness measured as 0, which is also not realistic.**

```python
# See how many cases have 0 Glucose, BMI, or Blood Pressure

zero_skin = training_set["SkinThickness"] == 0

zero_skin_cases = training_set[zero_g]

print(len(zero_skin_cases), "cases with 0 Skin Thickness")
```

---

**For cases of 0 Glucose, BMI, or Blood Pressure:**

Imputing a mean or median into these flawed cases might skew the data a little.

Since there are only 38 cases, I will remove these rows.

---


**For cases of 0 Skin Thickness:**

With 182 cases of 0 skin thickness, it might be too much to remove.

I will remove the feature's column instead.

```python
# create function with all cleaning steps
def clean_set(X_data, y_data):
    
    zero_g = X_data["Glucose"] == 0
    zero_bmi = X_data["BMI"] == 0
    zero_bp = X_data["BloodPressure"] == 0
    
    # get index of cases with 0 Glucose, BMI, or Blood Pressure
    indexes_to_remove = (zero_g | zero_bmi | zero_bp)
    
    # remove the zero cases
    X_data = X_data[~indexes_to_remove]
    y_data = y_data[~indexes_to_remove]
    
    # remove SkinThickness column
    X_data = X_data.drop("SkinThickness", axis=1)
    
    return X_data, pd.DataFrame(y_data)


# clean all the train and test sets together
cleaned_X_train, cleaned_y_train = clean_set(X_train, y_train)
cleaned_X_test, cleaned_y_test = clean_set(X_test, y_test)



# Prints to check cleaned sets ======================

print("\n\n")
print("cleaned_X_train.shape = ", cleaned_X_train.shape)
print("cleaned_X_test.shape = ", cleaned_X_test.shape)

print("cleaned_y_train.shape = ", cleaned_y_train.shape)
print("cleaned_y_test.shape = ", cleaned_y_test.shape)

print("\n")
print("cleaned_y_train class proportions: \n", cleaned_y_train.value_counts(normalize=True))

print("\n")
print("cleaned_y_test class proportions: \n", cleaned_y_test.value_counts(normalize=True))
```
